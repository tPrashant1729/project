```python
import pandas as pd
import sklearn as sk
```

```python
df = pd.read_csv("./WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
df.info()
```

```python
df.Churn.value_counts()
```

```python
df.dropna(inplace=True)
```

```python
df.replace({"Yes":1, "No":0},inplace = True)
```

```python
df.sample(5)
```

```python
df.replace({"Female":0,
            "Male":1,
            "No phone service":0,
            "No internet service":0}, inplace = True)
```

```python
df.info()
```

```python
df.InternetService.value_counts()
```

```python
df[df.InternetService == 0]
```

```python
df.replace({"Fiber optic":1,
            "DSL":1,
            "0":0 }, inplace = True)
```

```python
df.Contract.value_counts()
```

```python
df.PaymentMethod.value_counts()
```

```python
df.isna().sum()
```

```python
dummies = pd.get_dummies(df[["PaymentMethod","Contract"]], drop_first=True)
dummies.head()
```

```python
df_dummies = pd.concat([df,dummies],axis = 1)
df_dummies.sample(5)
```

```python
df_dummies.info()
```

```python
df_dummies.drop(["PaymentMethod","Contract","customerID"],axis=1, inplace = True)
```

```python
df_dummies.dtypes
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df_dummies.drop(["Churn"], axis=1)
y = df_dummies["Churn"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1, stratify=y)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_test_scaled, y_test)
```

```python
model.get_params()
```

```python
pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X_train_scaled, y_train, cv=5)
print(scores)
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
scores = cross_val_score(logistic, X_train_scaled, y_train, cv=5)
print(scores)

```

```python
logistic.fit(X_test_scaled, y_test)
pred = logistic.predict(X_test_scaled)
print(classification_report(y_test, pred))
```

```python
from sklearn import svm 
svm_clf = svm.SVC()
scores = cross_val_score(svm_clf, X_train_scaled, y_train, cv=5)
scores
```

```python
svm_clf.fit(X_test_scaled, y_test)
pred = svm_clf.predict(X_test_scaled)
print(classification_report(y_test, pred))
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'kernel':('linear', 'rbf'), 'C':[1, 10]}
svc = svm.SVC()
clf = GridSearchCV(svc, parameters,n_jobs=-1)
clf.fit(X_train_scaled,y_train)
```
